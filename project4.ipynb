```python
# 1. 2 players = 2 functions (done)

# 2. randomize 2 functions https://stackoverflow.com/questions/4859292/how-to-get-a-random-value-from-dictionary
# random.choice([d.keys()]).  or import random

# 3. randomize 2 seperate dictionaries for the cards of the players
# start with dict, add in all cards, after randomize, create list to keep values only

# 4. player, dealer, hit, stand

# 5. start 52 cards = list. 4 lists of 13 
# 4 diff suits:Clubs, Diamonds, Hearts and Spades
# For each suit, there will be cards #1-3

#Note: No wildcards will be used in the program, only 13 cards if u want to use

# 6. game begins: dealer will shuffle the deck of cards, making them randomized. 

# 7. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards. 

# 8. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.

# 9. objective= Player counts their cards after they're dealt. If not satisfied with the number, they have the ability to 'Hit'. 

# 10. A hit allows the dealer to deal the Player one additional card. )

# 11. The Player can hit as many times as they'd like as long as they don't 'Bust'. (A bust is when the Player is dealt cards that total >21.)

#12. If dealer deals the Player cards = 21 on the first deal, the Player wins. This is referred to as Blackjack. Blackjack is NOT the 
#same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

#import modules and define variables

import random #to pick out random cards from a deck

suits = ['clubs', 'spades', 'hearts', 'diamonds']
ranks = ['two','three','four','five','six','seven','eight','nine','ten', 'jack', 'queen', 'king', 'ace']
values = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 4, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10, 'jack': 10, 'queen': 10, 'king': 10, 'ace': 11}
#we put values in dict so we can map to them when creating our deck, picking them out, and hitting/standing

#we are now going to define another variable called playing

playing = True #when game is playing properly, only when player is actually doing stuff

#now we are going to define some CLASSES

class Card: #Creates all the cards 

    #after creating a class we want to add a few attributes

    def __init__(self,suit,rank): #suit and rank two variables that we are going to create our card names out of
        self.suit = suit
        self.rank = rank
    
    #__.__= magic methods, we are using string magic method below. what this method does it converts into a str
    #but only used in classes
    def __str__(self):
        return self.rank + ' of ' + self.suit 
        #this is going to return something like ex: three of clubs because rank is like 2


class Deck: 
    #what this class does it creates a deck of cards, so we do not 
    #need to manually do it and keep on shuffling it 
    def __init__(self):
        self.deck = [] #empty list because we have not created a deck yet
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank)) # 2 dots is normal especially in 2 classes

    def __str__(self):
        deck_comp = '' # empty str because we havent created our combo of cards
        for card in self.deck: #for every card in empty list line 66
            for card in self.deck: #for every card in empty list line 66
                deck_comp += '\n ' + card.__str__() #deck comp is = to, then new line and we concatenate that with card. then magic method
    #my deck_comp + a card, \n, and deck_comp = card. we are setting the value again and reasigning everytime we make new card
        return 'The deck has: ' + deck_comp #deck comp is the combination of cards in my deck


    def shuffle(self): #method here shuffles all the cards in the deck
        random.shuffle(self.deck)
    
    def deal(self): #pick out a card from the deck
        #we are creating another variable so we can store that card from shuffle
        single_card = self.deck.pop() #here we are saying the card we pick out is the card that is popped off from the end of the deck
        return single_card #after .pop() we return single card

        #^ there we created our deck method. this game will have quite a lot of classes

    #dealers and players hands 

class Hand: #show all the cards that the dealer and player have

    def __init__(self):
        self.cards = [] 
        self.value = 0 #total sum of values = 0
        self.aces = 0 # using this to keep track of aces, because we need aces to be in our game as well
    #add a method
    def add_card(self, card): #add a card to the players or dealers hand
        #add cards to our empty list
        self.cards.append(card) #self.cards reprpesent all cards in players or dealers hands
        self.value += values[card.rank] #here we are calculating the total sum of the values of the cards
        if card.rank == 'Ace':
            self.aces += 1 #add 1 to self.aces everytime the card picked out is an ace

    def adjust_for_ace(self):  #we are going to adjust for ace depending on the players total sum, if sum is close to 21 we make the ace turn to 1, if its 11 we are going to make it turn into 1
        while self.value > 21 and self.aces:
            self.value -=10 #minues 10 from total value if bigger than 21 and if we have an ace
            self.aces -= 1 # we can adjust ace if user is having a hard time

    #create another class 
class Chips: #this class keeps track of chips

    def __init__(self):
        self.total = 100 #when the user lost the amount of chips they bet was taken away from the self.total in this case 100 and there and the chips that the player received was 100- the amount of chips they bet
        self.bet = 0 #0 because player hasnt bet anything yet
    
    def win_bet(self):
        self.total += self.bet #total + the self.bet is =total, we are reassigning the total but after reassigning we are going ot add self.bet to it

    def lose_bet(self): #take away chips from 100 and return to you
        self.total -= self.bet 

#now we are going ot define our functions!

def take_bet(chips): #this is going to ask for users bet
    
    while True:
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
#using a try and accept here because if user types in words it will say sorry u have to type in anint
        except ValueError:
            print("Sorry! Please can you type in a number: ")
        else:
            if chips.bet > chips.total:
                print("Your bet can't exceed 100!")
            else:
                break

#if dealer or player wins they will get another cut

def hit(deck, hand): #add deck/hand because we all want to add to a hand and take some card from the deck
    hand.add_cards(deck.deal()) #use deal() function
    hand.adjust_for_ace() #incase it is an ace

def hit_or_stand(deck, hand): #what this function is going to do is ask if we want to hit or stance, it will ask player
    global playing #this means we are making this available in this function, will be looked at 1st

    while True: 
        ask = input("\nwould you like to hit or stand? Please enter 'h' or 's' ")

        if ask[0].lower() == 'h': #0 because in case they accidently type in something like hat, it will still count as hit
            hit(deck, hand)
        elif ask[0].lower() == 's':
            print("Player stands, dealer is playing")
            playing = False #this wont stop loop of our game, it will stop our player from doing anything else while the dealer is playing
        else:
            print("Sorry!, I did not understand that! Please try again! ")
            continue #continue this loop
        break  #it wont break this, it will break elif or else if we get what we wanted
        

def show_some(player, dealer): #showing player and dealers hands
    #sep special parameter that will seperate our cards apart
    print("\nDealer's Hand: ")
    print(" <card hidden>")
    print("", dealer.cards[1])
    print("\nPlayer's Hand: ", *player.cards, sep='\n ')

def show_all(player, dealer): #showing player and dealers hands
    #sep special parameter that will seperate our cards apart
    print("\nDealer's Hand: ", *dealer.cards, sep='\n')
    print("Dealer's Hand = ", dealer.value)
    print("\nPlayer's Hand: ", *player.cards, sep= '\n ')
    print("Player's Hand =", player.value)

#game ending, python interpreter will look at our game ending

def player_busts(player, dealer, chips): #if player busts we want to see chips
    print("PLAYER BUSTS! ")
    chips.lose_bet()


def player_wins(player, dealer, chips): #if player busts we want to see chips
    print("PLAYER WINS! ")
    chips.win_bet()


def dealer_busts(player, dealer, chips): #if player busts we want to see chips
    print("DEALER BUSTS! ")
    chips.win_bet()


def dealer_wins(player, dealer, chips): #if player busts we want to see chips
    print("DEALER WINS! ")
    chips.lose_bet()


def push(player, dealer): #no one will get chips if we tie
    print("Its a push! Player and Dealer tie! ")


# our classes and function ^we are our building blocks
#gameplay, while loops!

while True:
    print("Welcome to Blackjack")

    #create and shuffle deck
    #if we display it will be awkward and take up a lot of space
    deck = Deck()
    deck.shuffle() #.shuffle to call shuffle method
# create player/dealers hands
    player_hand = Hand()
    player_hand.add_card(deck.deal()) #when we add cards we want to put deck.deal so we can take out a card
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())

    #setup the player's chips
    player_chips = Chips()

    #ask player for bet
    take_bet(player_chips)

    # show cards

    show_some(player_hand, dealer_hand)

#ask player to hit or stand
    while playing:

        hit_or_stand(deck, player_hand)
        show_some(player_hand, dealer_hand)

        if player_hand.value >21:
            player_busts(player_hand, dealer_hand, player_chips)
            break

    if player_hand.value <= 21:

        while dealer_hand.value < 17:
            hit(deck, dealer_hand)
        
        show_all(player_hand, dealer_hand)

        if dealer_hand.value > 21:
            dealer_busts(player_hand, dealer_hand, player_chips)
        
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand, dealer_hand, player_chips)

        elif dealer_hand.value < 21:
            player_wins(player_hand, dealer_hand, player_chips)
        
        if player_hand.value > 21:
            player_busts(player_hand, dealer_hand, player_chips)
        
    print("\nPlayer's winnings stand at" , player_chips.total)

    #new variable called new game to aks player if he wants to play again

    new_game = input("\nWould you like to play again? Enter 'y' or 'n': ")
    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print("\nThanks for playing!")
        break

```
